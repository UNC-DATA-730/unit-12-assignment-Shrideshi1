**Create and visualize your model below...**

````r
library(splines)
library(tidymodels)
rec <- recipe(mpg ~ displacement + cylinders, data = Auto) |>
  step_ns(displacement, deg_free = 4) |>        # spline transform
  step_mutate(cyl_cat = factor(cylinders)) |>    # convert to factor
  step_dummy(cyl_cat) |>                          # dummy variables
  step_rm(cylinders)              

````

````r
mod <- linear_reg() |>
  set_engine("lm")

wf <- workflow() |>
  add_recipe(rec)|>
  add_model(mod)


mod_fit <- fit(wf, data = Auto)
````

````r
rec <- recipe(mpg ~ displacement + cylinders, data = Auto) %>%
  step_ns(displacement, deg_free = 4) %>% 
  step_mutate(cyl_cat = factor(cylinders)) %>%
  step_dummy(cyl_cat) %>%
  step_rm(cylinders)

# Prep with retain = TRUE
rec_prep <- prep(rec, retain = TRUE)

# Juiced data
train_data <- juice(rec_prep)

# Add original variables BACK for plotting
train_data$displacement <- Auto$displacement
train_data$cylinders <- Auto$cylinders

# Fit model
mod <- linear_reg() |> set_engine("lm")
mod_fit <- mod |> fit(mpg ~ ., train_data)

# Add predictions
train_data$.pred <- predict(mod_fit, train_data)$.pred

# Plot
ggplot(train_data, aes(x = displacement, y = mpg)) +
  geom_point(size = 1, alpha = 0.35) +
  geom_line(aes(y = .pred, color = factor(cylinders)), linewidth = 1) +
  labs(
    title = "MPG vs Displacement (Spline Model with Cylinder Categories)",
    x = "Displacement",
    y = "MPG",
    color = "Cylinders"
  ) +
  theme_minimal()
````
